```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("Position_Salaries.csv")
X = dataset.iloc[:, 1:-1].values
Y = dataset.iloc[:, -1].values
```

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X, Y)
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly_reg = PolynomialFeatures(degree=4)
X_poly = poly_reg.fit_transform(X)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, Y)
```

```python
Y_pred_lin = lin_reg.predict(X)
plt.scatter(X, Y, color="red")
plt.title("Truth or Bluff (LR)")
plt.plot(X, Y_pred_lin)
plt.xlabel("Position Level")
plt.ylabel("Salary")
```

```python
Y_pred_lin_2 = lin_reg_2.predict(X_poly)
plt.scatter(X, Y, color="red")
plt.title("Truth or Bluff (PR)")
plt.plot(X, Y_pred_lin_2)
plt.xlabel("Position Level")
plt.ylabel("Salary")
```

```python
var = np.array([6.5]).reshape(-1, 1)
pred_lin = lin_reg.predict(var)
pred_lin_2 = lin_reg_2.predict(poly_reg.fit_transform(var))
```

```python
Y_pred_lin_2 = lin_reg_2.predict(X_poly)
plt.scatter(X, Y, color="red")
plt.title("Truth or Bluff (PR)")
plt.plot(X, Y_pred_lin_2, color="blue")
plt.plot(X, Y_pred_lin, color="orange")
plt.xlabel("Position Level")
plt.ylabel("Salary")

plt.scatter(var, pred_lin, color="yellow")

plt.scatter(var, pred_lin_2, color="green")
```
