```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("Position_Salaries.csv")
X = dataset.iloc[:, 1:-1].values
Y = dataset.iloc[:, -1].values
```

```python
from sklearn.ensemble import RandomForestRegressor

rfr_reg = RandomForestRegressor(n_estimators=50000)
rfr_reg.fit(X, Y)
```

```python
Y_pred = rfr_reg.predict(X)
```

```python
X_grid = np.arange(min(X), max(X), 0.00001)
X_grid = X_grid.reshape(len(X_grid), 1)
plt.scatter(X, Y, color="red")
plt.plot(X_grid, rfr_reg.predict(X_grid), color="blue")
plt.title("High Resolution Decision Tree Visualisation")
plt.xlabel("Position Level")
plt.ylabel("Salary")
```
